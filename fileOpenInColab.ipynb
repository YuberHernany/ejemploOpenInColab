## Ejemplo de creación de una clase Persona en Python

```python
class Persona:
    def __init__(self, nombre, edad):
        """inputs: 
        nombre: str
        edad: int """
        self.nombre = nombre
        self.edad = edad
    def __str__(self):
        return str(self.nombre)
    def esMayorQue(self, otra):
        return self.edad > otra.edad
    def esMayorDeEdad(self):
        return self.edad > 18
```

Ahora instanciemos personas y ejemplifiquemos algunos cálculos

```python
yuber = Persona("Yuber Tapias", 37)
lina = Persona("Lina Gomez", 37)

print(yuber)
print(yuber.esMayorDeEdad())
print(yuber.esMayorQue(lina))
```
